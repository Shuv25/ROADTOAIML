```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:\\DATASETS\\heart.csv")
df.head()
```

```python
df['sex']=df['sex'].map({1:'male',0:'female'})
```

```python
df.head(3)
```

```python
sns.scatterplot(x='age',y='chol',hue='sex',style='sex',data=df)
plt.xlabel("Age")
plt.ylabel("Cholestrol")
plt.title("Age Vs Chol")
plt.show()
```
