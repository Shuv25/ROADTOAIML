Created a Function for all bar plots using matplot

```python
import random #for randomization

color_palettes = ['magma', 'viridis', 'plasma', 'inferno', 'cividis', 'Blues', 'Greens', 'Oranges', 'Purples', 'Reds']#different types of color palettes in seaborn

def plot_column_distribution(df, column):
    
    unique_values = df[column].unique() #finding the unique values present inside the column
    value_counts = df[column].value_counts()#counting each value and sum them
    

    chosen_palette = random.choice(color_palettes) #will select a random color pallette
    
    colors = sns.color_palette(f"{chosen_palette}_r", len(unique_values)) #will generate no of color based on the len of unique values and _r used as reverse of the color palette
    
    plt.bar(unique_values, value_counts, color=colors, label=unique_values)#bar plot 
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.title(f"Count of {column}")
    plt.legend(title=column)
    plt.show()
```

```python
def plot_with_random_palette(df, col1, col2):
    color_palettes = ['viridis', 'plasma', 'inferno', 'magma', 'cividis', 
                      'Blues', 'Greens', 'Purples', 'Reds', 'coolwarm', 
                      'Set1', 'Set2', 'Set3', 'Paired', 'Pastel1']
    
    chosen_palette = random.choice(color_palettes)
    
    sns.countplot(x=col1, hue=col2, data=df, palette=chosen_palette)#count plot
    plt.xlabel(col1)
    plt.ylabel("Count")
    plt.title(f"Count of {col1} by {col2} ")
    plt.show()
```

```python
def plot_pie_chart(df, col_name):
    color_palettes = ['viridis', 'plasma', 'inferno', 'magma', 'cividis', 
                      'Blues', 'Greens', 'Purples', 'Reds', 'coolwarm', 
                      'Set1', 'Set2', 'Set3', 'Paired', 'Pastel1']
  
    chosen_palette = random.choice(color_palettes)
    

    plt.figure(figsize=(5, 5))
    
    value_counts = df[col_name].value_counts()
    
    value_counts.plot(kind='pie', autopct='%1.1f%%', 
                      colors=sns.color_palette(chosen_palette, len(value_counts)))#pie plot
    
    plt.title(f'Pie Chart of {col_name}')
    plt.show()
```

```python
plot_pie_chart(df,'Age_Group')
```
