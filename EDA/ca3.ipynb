```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Credit amount',axis=1),df['Credit amount'],test_size=0.3,random_state=56)
```

```python
num_col=[i for i in X_train if X_train[i].dtype!='O']
```

```python
cat_col=[i for i in X_train if X_train[i].dtype =='O']
```

```python
num_pipe=Pipeline(steps=[
    ('scaler',StandardScaler())
])
cat_pipe=Pipeline(steps=[
    ('encoder',OneHotEncoder(sparse_output=False,handle_unknown='ignore'))
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('scaler',num_pipe,num_col),
    ('encoder',cat_pipe,cat_col),
])
```

```python
model=Pipeline(steps=[
    ('pre',preprocessor),
    ('lr',LinearRegression())
])
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("Accuracy:",np.round(r2_score(y_test,y_pred),2))
```
