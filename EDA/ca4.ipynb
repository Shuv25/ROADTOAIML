```python
#here i am creating a function to plot bar chart on each col
import random

color_palletes=['magma', 'viridis', 'plasma', 'inferno', 'cividis', 'Blues', 'Greens', 'Oranges', 'Purples', 'Reds']#this are the seaborn inbuilt colourmaps

def barplot(df,col):
    
    chosen_palette=random.choice(color_palletes)#choosing any colour pallete randomly with the help of random library

    unique_names=df[col].value_counts().index#saving the unique value index
    unique_values=df[col].value_counts()#all the value of unique value index

    colors=sns.color_palette(f"{chosen_palette}_r",len(unique_names))#storing the chosen color on colors variable

    if(len(unique_names)<15):
        plt.figure(figsize=(9,6))
        plt.bar(unique_names,unique_values,color=colors,label=unique_names)
        plt.title(f"Bar plot of {col}")
        plt.xlabel(col)
        plt.ylabel(f"Count of {col}")
        plt.legend(title=col)
        plt.show()
    else:
        plt.figure(figsize=(15,7))
        unique_values=df[col].value_counts()[:10]
        unique_names=df[col].value_counts().index[:10]
        plt.figure(figsize=(13,7))
        plt.bar(unique_names,unique_values,color=colors,label=unique_names)
        plt.title(f"Bar plot of Top 10 {col}")
        plt.xlabel(col)
        plt.ylabel(f"Count of {col}")
        plt.legend(title=col)
        plt.xticks(rotation=90)
        plt.show()
```

```python
for i in df.columns:
        barplot(df,i)#calling the function inside a loop
```

```python
#here i am also creating count plot instead of bar and everything is same as previous
color_palletes=['magma', 'viridis', 'plasma', 'inferno', 'cividis', 'Blues', 'Greens', 'Oranges', 'Purples', 'Reds']

def count_plot(df,col_one,col_two):
    
    chosen_palette=random.choice(color_palletes)

    if(len(df[col_two].unique())<15):
        plt.figure(figsize=(10,7))
        sns.countplot(x=col_one,hue=col_two,data=df,palette=chosen_palette)
        plt.title(f"Count plot of {col_one} and {col_two}")
        plt.xlabel(col_one)
        plt.ylabel(f"Count")
        plt.show()
    else:
        top_values = df[col_two].value_counts().index[:10]
        df_top = df[df[col_two].isin(top_values)]

        plt.figure(figsize=(15,7))
        sns.countplot(x=col_one,hue=col_two,data=df_top,palette=chosen_palette)
        plt.title(f"Count plot of Top 10 {col_one} and {col_two}")
        plt.xlabel(col_one)
        plt.ylabel(f"Count")
        plt.xticks(rotation=90)
        plt.show()
```

```python
for i in df.columns:
    if(i!='Age_Group'):
        count_plot(df,'Age_Group',i)
```

```python
#here i am also creating hist instead of bar and everything is same as previous
color_palletes=['magma', 'viridis', 'plasma', 'inferno', 'cividis', 'Blues', 'Greens', 'Oranges', 'Purples', 'Reds']

def hist_plot(df,col_one,col_two):
    
    chosen_palette=random.choice(color_palletes)

    if(len(df[col_two].unique())<15):
        plt.figure(figsize=(10,7))
        sns.histplot(x=col_one,hue=col_two,data=df,palette=chosen_palette+"_r")
        plt.title(f"Hist plot of {col_one} and {col_two}")
        plt.xlabel(col_one)
        plt.ylabel(f"Count")
        plt.show()
    else:
        top_values = df[col_two].value_counts().index[:10]
        df_top = df[df[col_two].isin(top_values)]

        plt.figure(figsize=(15,7))
        sns.histplot(x=col_one,hue=col_two,data=df_top,palette=chosen_palette+"_r")
        plt.title(f"Hist plot of Top 10 {col_one} and {col_two}")
        plt.xlabel(col_one)
        plt.ylabel(f"Count")
        plt.xticks(rotation=90)
        plt.show()
```

```python
#here i am also creating pie plot instead of bar and everything is same as previous
def pie_chart(df, col):
    color_palettes = ['viridis', 'plasma', 'inferno', 'magma', 'cividis', 
                      'Blues', 'Greens', 'Purples', 'Reds', 'coolwarm', 
                      'Set1', 'Set2', 'Set3', 'Paired', 'Pastel1']
  
    chosen_palette = random.choice(color_palettes)
    
    value_counts = df[col].value_counts()
    
    if(len(value_counts)<10):
        plt.figure(figsize=(5, 5))
        value_counts.plot(kind='pie', autopct='%1.1f%%', 
                      colors=sns.color_palette(chosen_palette, len(value_counts)))
        plt.title(f'Pie Chart of {col}')
        plt.show()
    else:
        value_counts = df[col].value_counts()[:10]
        plt.figure(figsize=(7, 9))
        value_counts.plot(kind='pie', autopct='%1.1f%%', 
                      colors=sns.color_palette(chosen_palette, len(value_counts)))
        plt.title(f'Pie Chart of {col}')
        plt.show()
```

```python
for i in df.columns:
    pie_chart(df,i)
```

```python
#importing all the necessary libraries for model classification
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Diet Satisfaction',axis=1),df['Diet Satisfaction'],test_size=0.3,random_state=56)#spliting dataset
```

```python
X_train
```

```python
num_cols=['Health Rating']#collecting the  num col names
cat_cols=[i for i  in X_train if X_train[i].dtype=='object']#collecting the cat col names
```

```python
print(num_cols)
print(cat_cols)
```

```python
#creating a pipline
num_pipe=Pipeline(steps=[
    ('scaler',StandardScaler())
])
cat_pipe=Pipeline(steps=[
    ('encoder',OneHotEncoder(sparse_output=False,handle_unknown='ignore'))
])
```

```python
#creating column transformer for not using repetative code
preprocessor=ColumnTransformer(transformers=[
    ('scaler',num_pipe,num_cols),
    ('encoder',cat_pipe,cat_cols),
])
```

```python
#here i am creating DecisionTreeClassifier classification model
model=Pipeline(steps=[
    ('pre',preprocessor),
    ('dtc',DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=None))
])
```

```python
model.fit(X_train,y_train)#fiting the model
```

```python
y_pred=model.predict(X_test)#storing the model pred in y_pred of X_test
```

```python
import numpy as np
```

```python
print("Accuracy:",np.round(accuracy_score(y_test,y_pred),2))#checking the accuracy of our model
```
