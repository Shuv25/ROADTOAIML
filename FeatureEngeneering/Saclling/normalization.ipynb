```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:\\DATASETS\\wine_data.csv")
df.head()
```

```python
df=df.iloc[:,:3]
```

```python
df.head()
```

```python
col_names=['Level','Alcohol','Malic_Acid']
df.columns=col_names
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Level',axis=1),df['Level'],test_size=0.2,random_state=56)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train_scalled=scaler.transform(X_train)
X_test_scalled=scaler.transform(X_test)
```

```python
X_train_scalled=pd.DataFrame(X_train_scalled,columns=X_train.columns)
```

```python
len(X_train_scalled)
```

```python
X_test_scalled=pd.DataFrame(X_test_scalled,columns=X_test.columns)
```

```python
np.round(X_train.describe(),2)
```

```python
np.round(X_train_scalled.describe(),2)
```

```python
len(X_train_scalled)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr_scalled=LinearRegression()
```

```python
lr.fit(X_train,y_train)
lr_scalled.fit(X_train_scalled,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred_scalled=lr_scalled.predict(X_test_scalled)
```

```python
from sklearn.metrics import r2_score
```

```python
print("Acuuracy:",r2_score(y_test,y_pred))
print("Acuuracy:",r2_score(y_test,y_pred_scalled))
```
