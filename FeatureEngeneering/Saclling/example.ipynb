```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("D:\\DATASETS\\insurance.xlsx")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['bmi'].fillna(df['bmi'].median(),inplace=True)
df['charges'].fillna(df['charges'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(['charges'],axis=1),df['charges'],test_size=0.2,random_state=56)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
```

```python
X_train_scalled=sc.transform(X_train)
X_test_scalled=sc.transform(X_test)
```

```python
X_train_scalled=pd.DataFrame(X_train_scalled,columns=X_train.columns)
X_test_scalled=pd.DataFrame(X_test_scalled,columns=X_test.columns)
```

```python
round(X_train.describe(),2)
```

```python
round(X_train_scalled.describe(),2)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm_scalled=LinearRegression()
```

```python
lm.fit(X_train,y_train)
lm_scalled.fit(X_train_scalled,y_train)
```

```python
model=lm.predict(X_test)
model_scalled=lm_scalled.predict(X_test_scalled)
```

```python
from sklearn.metrics import r2_score
```

```python
print("Normal",r2_score(y_test, model))
print("Scalled",r2_score(y_test, model_scalled))
```
