KMeans Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:\\DATASETS\\student_clustering.csv")
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['cgpa'],df['iq'])
plt.title("Scatter Plot of CGPA and IQ")
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(wcss)
plt.title("Elbo Curve")
plt.xlabel("No of Clusters")
plt.ylabel("Inrecia or WCSS")
plt.show()
```

```python
X=df.iloc[:,:].values
km=KMeans(n_clusters=4)
y_means=km.fit_predict(X)
```

```python
y_means
```

```python
X[y_means==0,0]#x cordinates of cluster 0
```

```python
X[y_means==0,1]#y cordinates of cluster 1
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],color='blue')
plt.scatter(X[y_means==1,0],X[y_means==1,1],color='red')
plt.scatter(X[y_means==2,0],X[y_means==2,1],color='green')
plt.scatter(X[y_means==3,0],X[y_means==3,1],color='yellow')
plt.title("Clusterd Plot")
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.show()
```
