```python
def diff_demo(my_f, my_x, my_delta):
    return (my_f(my_x + my_delta) - my_f(my_x)) / my_delta#here mydelta is x2-x1 
```

```python
deltas = [1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]
```

```python
for delta in deltas:
    print(diff_demo(yfun, 2, delta))
```

```python
for delta in deltas:
    print(diff_demo(yfun, -1, delta))
```
