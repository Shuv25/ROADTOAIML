```python
def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each. 

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]] 
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
        
    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

```python
plot_vectors([Av,v,Av2,v2],['blue','lightblue','green','lightgreen'])
plt.xlim(-1,4)
_ = plt.ylim(-2,4)
```

Image Compression Using SVD

```python
from PIL import Image
import requests
from io import BytesIO
```

```python
url = "https://www.enjpg.com/img/2020/gojo-satoru-scaled-e1605384826807.jpg"
response = requests.get(url)
```

```python
img = Image.open(BytesIO(response.content))
_ = plt.imshow(img)
```

Convert image to grayscale so that we don't have to deal with the complexity of multiple color channels:

```python
imgrey=img.convert('LA')
_ =plt.imshow(imgrey)#this image is a 2d vector, in this image we have 2000 row , 2500 column
```

```python
imgmat = np.array(list(imgrey.getdata(band=0)), float)
imgmat.shape = (imgrey.size[1], imgrey.size[0])
imgmat = np.matrix(imgmat)
_ = plt.imshow(imgmat, cmap='gray')
```

```python
U,sigma,V=np.linalg.svd(imgmat)
```

As eigenvalues are arranged in descending order in diag(
) so too are singular values, by convention, arranged in descending order in 
 (or, in this code, diag(
)). Thus, the first left-singular vector of 
 and first right-singular vector of 
 may represent the most prominent feature of the image:

```python
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
_ = plt.imshow(reconstimg, cmap='gray')
```

```python
for i in [2, 4, 8, 16, 32, 64]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

Solving Regression Problem

```python
x1 = [0, 1, 2, 3, 4, 5, 6, 7.] # E.g.: Dosage of drug for treating Alzheimer's disease
y = [1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37] # E.g.: Patient's "forgetfulness score"
```

```python
title = 'Clinical Trial'
xlabel = 'Drug dosage (mL)'
ylabel = 'Forgetfulness'
```

```python
fig, ax = plt.subplots()
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
_ = ax.scatter(x1, y)
```

```python
x0 = np.ones(8)
x0
```

```python
X = np.concatenate((np.matrix(x0).T, np.matrix(x1).T), axis=1)
X
```

```python
w=np.dot(np.linalg.pinv(X),y)
w
```

```python
b = np.asarray(w).reshape(-1)[0]
b
```

```python
m=np.asarray(w).reshape(-1)[1]
m
```

```python
fig, ax = plt.subplots()

plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)

ax.scatter(x1, y)

x_min, x_max = ax.get_xlim()
y_at_xmin = m*x_min + b
y_at_xmax = m*x_max + b

ax.set_xlim([x_min, x_max])
_ = ax.plot([x_min, x_max], [y_at_xmin, y_at_xmax], c='C01')
```
