```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from pandas import Series
from random import randrange
```

```python
series=[i+randrange(10) for i in range(1,100)]
```

```python
series
```

Additative

```python
result=seasonal_decompose(series,model='additive',period=1)
result.plot()
plt.show()
```

Multiplicative

The model here is Multiplicative

The period corresponds to the frequency of the data, meaning how many data points make up one season or cycle.
For example:
Daily data with yearly seasonality: If the data is collected daily, and you expect yearly seasonality, the period would be 365.
Monthly data with yearly seasonality: If the data is monthly, the period would be 12.
Quarterly data with yearly seasonality: The period would be 4.
Weekly data with yearly seasonality: If the data is weekly, the period might be 52 (for 52 weeks in a year).

```python
result_2=seasonal_decompose(series,model='multiplicative',period=1)
result_2.plot()
plt.show()
```
