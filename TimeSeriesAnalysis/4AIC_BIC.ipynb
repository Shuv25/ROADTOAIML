```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA 
```

```python
df=pd.read_csv("D:\\DATASETS\\monthly_beer.csv")
df.head()
```

```python
df.set_index('Month',inplace=True)
df.index=pd.to_datetime(df.index)
df.dropna(inplace=True)
```

```python
df.plot()
plt.show()
```

```python
sd=seasonal_decompose(df['Monthly beer production'],model='multiplicative',period=12)
sd.plot()
plt.show()
```

```python
ADF=adfuller(df)
ADF
```

```python
plot_acf(df,lags=12)
plt.show()
```

```python
plot_pacf(df,lags=12)
plt.show()
```

```python
df_diff=df['Monthly beer production'].diff()
```

```python
df_diff = df_diff.dropna()
```

```python
df_diff
```

```python
from itertools import product

for p in range(1, 12):
    try:
        arma_model = ARIMA(df_diff, order=(p, 0, 0))
        arma_model_fit = arma_model.fit()
        aic_score = arma_model_fit.aic
        bic_score = arma_model_fit.bic
        print(f"ARIMA({p},0,0) - AIC: {aic_score},BIC: {bic_score}")
    except Exception as e:
        print(f"Error for ARIMA({p},0,0): {e}")
```

```python
train=df_diff[0:333]
test=df_diff[333:]
print(len(train))
print(len(test))
```

```python
model=ARIMA(train,order=(11,0,0))
```

```python
model_fit=model.fit()
```

```python
print(model_fit.summary())
```

```python
df.tail()
```

```python
start_date_index="1995-09-01"
end_date_index="1995-12-01"
```

```python
forcast=model_fit.predict(start=start_date_index,end=end_date_index)
```

```python
forcast
```

```python
plt.figure(figsize=(12,6))
plt.plot(test)
plt.plot(forcast)
plt.legend(("Test Data","Prediction"))
plt.show()
```

```python
def invert_diff(last_actual,differences):
    inverted_values=[]
    last_observation=last_actual
    for diff in differences:
        inverted_value=last_observation+diff
        inverted_values.append(inverted_value)
        last_observation=inverted_value
    return inverted_values
```

```python
forcast_og=invert_diff(df['Monthly beer production'].iloc[-1],forcast)
```
