```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread('gojo.jpg')
```

```python
plt.imshow(img1)
plt.show()
```

```python
img2=cv2.imread('gojo.jpg',cv2.IMREAD_GRAYSCALE)#we also have IMREAD_COLOR,IMREAD_UNCHANGED
```

```python
plt.imshow(img2,cmap='gray',interpolation='nearest')
plt.show()
```

```python
plt.imshow(img2,cmap='gray',interpolation='bilinear')
plt.show()
```

```python
plt.imshow(img2,cmap='gray',interpolation='lanczos')
plt.show()
```

```python
img3=cv2.imread('gojo.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img3,cmap='gray',interpolation='nearest')
plt.plot([100,500],[190,500],'r',linewidth=5)#here i ma drwing a line on the image first list contains x1,y1 second list is x2,y2 and r is color red and atlast linewidth is width of the line
plt.show()
```

```python
cv2.imwrite('graygojo.jpg',img3)
```
