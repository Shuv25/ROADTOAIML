```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(train.drop('loan_status',axis=1),train['loan_status'],test_size=0.2,random_state=56)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oc=OneHotEncoder(drop='first',dtype=np.int32)
X_train_new=oc.fit_transform(X_train[['person_home_ownership','loan_intent','loan_grade','cb_person_default_on_file']]).toarray()
X_test_new=oc.transform(X_test[['person_home_ownership','loan_intent','loan_grade','cb_person_default_on_file']]).toarray()
```

```python
encoded_feature_names = oc.get_feature_names_out(['person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file'])
```

```python
encoded_feature_names
```

```python
X_train_encoded_df = pd.DataFrame(X_train_new, columns=encoded_feature_names)
```

```python
X_test_encoded_df = pd.DataFrame(X_test_new, columns=encoded_feature_names)
```

```python
X_train_encoded_df.shape
```

```python
X_train
```

```python
categorical_columns = ['person_home_ownership', 'loan_intent', 'loan_grade', 'cb_person_default_on_file']
X_train_without_categorical = X_train.drop(categorical_columns, axis=1)
```

```python
X_test_without_categorical = X_test.drop(categorical_columns, axis=1)
```

```python
X_train_final = pd.concat([X_train_without_categorical.reset_index(drop=True), X_train_encoded_df.reset_index(drop=True)], axis=1)
```

```python
X_train_final
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_test_final = pd.concat([X_test_without_categorical.reset_index(drop=True), X_test_encoded_df.reset_index(drop=True)], axis=1)
```

```python
X_test_final
```

```python
sc=StandardScaler()
sc.fit(X_train_final)
X_train_final_scaled=sc.transform(X_train_final)
X_test_final_scaled=sc.transform(X_test_final)
```

```python
X_train_scaled=pd.DataFrame(X_train_final_scaled,columns=X_train_final.columns)
X_test_scaled=pd.DataFrame(X_test_final_scaled,columns=X_test_final.columns)
```

```python
# X_train_scaled.drop('id',axis=1,inplace=True)
# X_test_scaled.drop('id',axis=1,inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
log.fit(X_train_scaled,y_train)
```

```python
X_test_scaled.shape
```

```python
y_pred=log.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
